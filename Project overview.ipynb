## Application Tracking System (ATS) with Machine Learning:

This project is a machine learning-based Application Tracking System (ATS) designed to help recruiters and companies automate the process of evaluating job applications and resumes. By utilizing natural language processing (NLP) and machine learning algorithms, the system analyzes resumes, identifies relevant information, and predicts the suitability of a candidate for a given job position.

The system integrates Google Generative AI (Gemini-1.5) for enhanced resume parsing and scoring. A web interface built using Streamlit allows users to upload PDFs, which are converted to images for easier processing and analysis.

### Problem Statement
Recruiters often receive hundreds or thousands of resumes for a single job opening, making it difficult and time-consuming to manually review all applications. The goal of this ATS project is to build an efficient, intelligent system that can:

- Automatically extract relevant information from resumes.
- Match candidates to job descriptions based on their skills and experiences.
- Provide recommendations for hiring decisions.

### Key Features

- Resume Analysis: Using Google Generative AI, the system parses resumes and extracts critical information such as skills, education, experience, and contact details.

- Document Processing: The system supports PDF uploads, converts them to images, and processes the images to extract text for analysis.

- Candidate Scoring: Based on the extracted features, candidates are scored and ranked according to their suitability for the job role.

- Streamlit Interface: A user-friendly web interface where recruiters can upload resumes, view candidate scores, and filter applicants.

### Data Collection

The system works with a dataset of resumes and job descriptions. Key attributes extracted from resumes include:

- Personal Information: Name, contact details, and location.
- Skills: A list of technical and soft skills.
- Experience: Job roles, companies worked at, and duration of each job.
- Education: Degrees obtained and institutions attended.
Job descriptions are analyzed similarly to identify the required qualifications and skills.


### Preprocessing

Before training the machine learning models, various preprocessing steps were performed to prepare the data:

- Text Cleaning: Removing special characters, numbers, and unnecessary symbols from resume text.
- Tokenization: Breaking down the text into words or tokens.
- Stop Words Removal: Common words that do not contribute to the context (e.g., "the", "is", "in") are removed.
- Feature Extraction: Extracting relevant features such as skills, qualifications, and experience from resumes using NLP techniques.

### Generative AI Integration

The project integrates Google Generative AI (Gemini-1.5) to generate candidate recommendations based on input data. This enhances the system's ability to analyze resumes, improve text extraction, and offer more human-like insights on candidate suitability.

### User Interface

The system was built with Streamlit, a Python library for creating interactive web applications. The interface allows users to:

- Upload resumes in PDF format.
- View the extracted details and scores of candidates.
- Get hiring recommendations based on machine learning predictions.

### Conclusion

This ATS project is designed to streamline the resume evaluation process using machine learning and AI techniques. With a user-friendly interface and powerful backend models, the system provides recruiters with intelligent candidate recommendations, helping reduce manual effort and improve hiring accuracy.

The integration of Google Generative AI and machine learning techniques ensures a robust solution for real-time resume analysis and hiring decisions.
